# **Create a DataFrame:**
**Create a DataFrame df from a dictionary containing the following
data:**

**data = {
 'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
 'Age': [25, 30, 35, 40, 45],
 'Salary': [50000, 60000, 70000, 80000, 90000]}**

**Basic DataFrame Operations**

**1)Display the first 3 rows of the DataFrame.**

**2)Display the summary statistics for numerical columns.**

**3)Retrieve the 'Age' column as a Series.**

**4)Calculate the average salary.**

```python
import pandas as pd
import numpy as np

data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],'Age': [25, 30, 35, 40, 45],'Salary': [50000, 60000, 70000, 80000, 90000]}

x = pd.DataFrame(data)

print(x)
```

```python
# 1. Display the first 3 rows of the DataFrame.

print(x[:3])
```

```python
# 2. Display the summary statistics for numerical columns.

print(x.describe())
```

```python
# 3. Retrieve the 'Age' column as a Series

age = pd.Series(x['Age'])
print(age)
```

```python
# 4. Calculate the average salary.

import numpy as np
print(np.mean(x["Salary"]))
```
